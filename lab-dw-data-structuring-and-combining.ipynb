In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd
customer_data=pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
customer_data2=pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
customer_data3=pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')
```

```python
sub_merge=pd.merge(customer_data,customer_data2,how='outer')
sub_merge.columns=sub_merge.columns.str.lower().str.replace(" ","_")
sub_merge.rename(columns={'st':'state'},inplace=True)
customer_data3.columns=customer_data3.columns.str.lower().str.replace(" ","_")
#Normalize gender column to M and F
dict_gender={'Male':'M','female':'F','Femal':'F','M':'M','F':'F','nan':'nan'}
sub_merge['gender']=sub_merge['gender'].map(dict_gender)
#Normalize state to initial.ex Arizona --> AZ
dict_state={'Washington':'WA', 'Arizona':'AZ', 'Nevada':'NV', 'California':'CA', 'Oregon':'OR', 'Cali':'CA','AZ':'AZ', 'WA':'WA','nan':'nan'}
sub_merge.state=sub_merge.state.map(dict_state)
#Remane Bachelors for Bachelor
sub_merge.loc[sub_merge['education']=='Bachelors','education']='Bachelor'
#Normalize Luxyry class in vehicule class
sub_merge.loc[ (sub_merge['vehicle_class'] == 'Sports Car') | (sub_merge['vehicle_class'] == 'Luxury SUV') | (sub_merge['vehicle_class'] == 'Luxury Car'),'vehicle_class'] ='Luxury'
#CLV delete the % character
sub_merge.customer_lifetime_value=sub_merge.customer_lifetime_value.str.replace("%"," ",)
#Convert CLV into numeric
sub_merge.customer_lifetime_value= sub_merge.customer_lifetime_value.astype(float)
#Fixe number of open complaints
dict_number_complaints={'1/0/00':"0", '1/2/00':"2", '1/1/00':"1", '1/3/00':"3", '1/5/00':"5", '1/4/00':"4",'nan':pd.NA}
sub_merge.number_of_open_complaints=sub_merge.number_of_open_complaints.map(dict_number_complaints).astype(float)
customer_data_final=pd.merge(sub_merge,customer_data3,how='outer')
customer_data_final
```
